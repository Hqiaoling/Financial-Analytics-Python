## E. What is dupont formula and analysis? 
The dupont analysis is a finacial ratio based on the return on assets ratio that is used to analyze a company's ability to increase its return on assets. 

There are three main conponents of ROA ratio:
1. Net Income
2. Revenues
3. Total Assets

**\*\*Dupont formula\*\***

The dupont formula looks like this:

$Return\:on\:Assets\:(ROA \%) = \frac{Net \: Income}{Assets} = \frac{Net\:Income}{Revenue} \times\frac{Revenues}{Assets} = Ratio\:(A) \times Ratio\:(B)$ 

**\*\*Dupont analysis\*\***

As we mentioned before, dupont analysis is used to analyze ROA and investors are looking to analyze what cause the current ROA increase or decrease. So investors can use this formula to analyze the problem based on the three main conponents: Net income, Revenues, Total assets. 

This Dupont analysis decomposes ROA into two ratio that provide a more detailed insight about how the company generate its ROA. As ratio(A) is profit margin ratio (PMR), it reflects the company's ability to operate effiency by controling its expenses and/or changing a price. We can also see how improving PMR can also lead to improve ROA. Ratio(B) is to measure the volume of sales revenues, this is also called Assets Turnover Ratio.






## F. Create a combined graph that plots the “Asset Turnover Ratio (ATR)” for the 5 companies from 2010-2018.

```
## calculate ATR and create a new column 'ATR' in data frame
df['ATR'] = df['Sales/Turnover (Net)'] / df['Assets - Total'] 
df.head()

## combine five companies' ATR into one data frame
Y_a1 = df.loc[0:8, 'ATR']
Y_a2 = df.loc[9:17, 'ATR'].reset_index(drop= True) 
Y_a3 = df.loc[18:26, 'ATR'].reset_index(drop= True) 
Y_a4 = df.loc[27:35, 'ATR'].reset_index(drop= True)
Y_a5 = df.loc[36:44, 'ATR'].reset_index(drop= True)
df_atr = pd.DataFrame({'X': X, 'Y1':Y_a1, 'Y2':Y_a2,'Y3':Y_a3,'Y4':Y_a4,'Y5':Y_a5})

## plot a combined graph of ATR for five companies from 2010 to 2018
plt.plot('X', 'Y1', data=df_atr, color='#c94818', linewidth=2, label = "DAL")
plt.plot('X', 'Y2', data=df_atr, color='#0a304d', linewidth=2, label = "KR")
plt.plot('X', 'Y3', data=df_atr, color='#84a39e', linewidth=2, label = "MSFT")
plt.plot('X', 'Y4', data=df_atr, color='#ffc110', linewidth=2, label = "SBUX")
plt.plot('X', 'Y5', data=df_atr, color='#a96529', linewidth=2, label = "AMZN")
plt.xlabel('Year', fontsize = 14)
plt.ylabel('ATR',fontsize = 14)
plt.title('Dynamic of ATR', fontsize = 14)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

## G. Create a combined graph that plots the “Profit Margin Ratio (PMR)” for the 5 companies from 2010-2018.

```
## calcalte PMR and create a new column 'PMR' in data frame
df['PMR'] = df['Net Income (Loss)'] / df['Sales/Turnover (Net)']
df.head()

## combine five companies' PMR into one data frame
Y_p1 = df.loc[0:8, 'PMR']
Y_p2 = df.loc[9:17, 'PMR'].reset_index(drop= True) 
Y_p3 = df.loc[18:26, 'PMR'].reset_index(drop= True) 
Y_p4 = df.loc[27:35, 'PMR'].reset_index(drop= True)
Y_p5 = df.loc[36:44, 'PMR'].reset_index(drop= True)
df_pmr = pd.DataFrame({'X': X, 'Y1':Y_p1, 'Y2':Y_p2,'Y3':Y_p3,'Y4':Y_p4,'Y5':Y_p5})

## plot a combined graph of PMR for five companies from 2010 to 2018
plt.plot('X', 'Y1', data=df_pmr, color='#c94818', linewidth=2, label = "DAL")
plt.plot('X', 'Y2', data=df_pmr, color='#0a304d', linewidth=2, label = "KR")
plt.plot('X', 'Y3', data=df_pmr, color='#84a39e', linewidth=2, label = "MSFT")
plt.plot('X', 'Y4', data=df_pmr, color='#ffc110', linewidth=2, label = "SBUX")
plt.plot('X', 'Y5', data=df_pmr, color='#a96529', linewidth=2, label = "AMZN")
plt.xlabel('Year', fontsize = 14)
plt.ylabel('PMR',fontsize = 14)
plt.title('Dynamic of PMR', fontsize = 14)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
